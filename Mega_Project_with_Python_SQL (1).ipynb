```python
#Task 6 :-
#Questions for Visualizations:-
#1)	How can you use matplotlib to create a bar chart showing the top 10 universities by their scores in 2012?
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/WorldUniversity.csv')
top_10_univ_2012 = data[data['year'] == 2012].nlargest(10, 'score')
University_names = top_10_univ_2012['institution']
score = top_10_univ_2012['score']

plt.figure(figsize =(10 ,6))
plt.barh(University_names , score  , edgecolor = 'black' , color = 'skyblue')
plt.title("Top 10 Universities by Score in 2012")
plt.ylabel("Institution")
plt.xlabel("Scores")
plt.gca().invert_yaxis()
plt.show()
```

```python
#3)	How can you use Altair to create a scatter plot comparing the quality of faculty and alumni employment for all universities in 2013?
import pandas as pd
import altair as alt

# Assuming you have loaded your dataset into a pandas DataFrame called 'df'
# Filter the data for the year 2013
df = pd.read_csv('/content/WorldUniversity.csv')
data_2013 = df[df['year'] == 2013]

# Create a scatter plot using Altair
scatter_plot = alt.Chart(data_2013).mark_circle(size=60).encode(
    x='quality_of_faculty',
    y='alumni_employment',

    tooltip=['institution' ,'quality_of_faculty', 'alumni_employment']
).properties(
    width=600,
    height=400,
    title='Scatter Plot: Quality of Faculty vs Alumni Employment (2013)'
)
scatter_plot
```

```python
#6)	Create a stacked area plot using Altair to represent the change in scores for the top 5 universities from 2012 to 2014.
import pandas as pd
import altair as alt

df = pd.read_csv('/content/WorldUniversity.csv')
top5 = df[df['year']== 2012].nlargest(5 , 'score')
top_5_Univ = top5['institution']
top5_data = df[(df['year'].between(2012 ,2014)) & (df['institution'].isin(top_5_Univ))]

area_plot = alt.Chart(top5_data).mark_area().encode(
    x = 'year',
    y =alt.Y('score', stack = 'normalize'),
    color = 'institution',
    tooltip = ['year' , 'score' , 'institution']
).properties(
    height = 600,
    width = 400,
    title = ('change in scores for the top 5 universities from 2012 to 2014')
)
area_plot
```

```python
#13)	How can you use matplotlib to create a stacked bar chart illustrating the distribution of universities in different countries in 2012?

import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/WorldUniversity.csv')
df_2012 = df[df['year']== 2012]
country_counts = df_2012['country'].value_counts()
plt.figure(figsize = (10,6))
plt.bar(country_counts.index, country_counts.values)
plt.xlabel('Country')
plt.ylabel('Number of universities')
plt.title('Distribution of universities in different countries in 2012')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
#14)	Create a heatmap using seaborn to visualize the correlation matrix of the numerical columns in the dataset
import pandas as pd
import seaborn as sns
df = pd.read_csv('/content/WorldUniversity.csv')
df = df.select_dtypes(include = ['float64' , 'int64'])
corr_df = df.corr()
plt.figure(figsize=(10,8))
sns.heatmap(data = corr_df , annot = True , cmap = 'coolwarm' , fmt = '.2f')
plt.title("Correlation matrix of the numerical columns")
plt.show()
```

```python
#15)	How can you use Altair to create a line chart showing the trend of scores for the top 5 universities in 2014

import pandas as pd
import altair as alt
df = pd.read_csv('/content/WorldUniversity.csv')
top_5_universities = df[df['year']== 2014].nlargest(5,'score')
line_chart = alt.Chart(top_5_universities).mark_line().encode(
    x= 'year',
    y= 'score',
    color = 'institution',
    tooltip= ['institution', 'score', 'year']
).properties(
    height = 600,
    width = 400,
    title = (' Trend of scores for the top 5 universities in 2014')
)
line_chart
```

```python
#16)	Using matplotlib, create a bar chart to show the average scores for universities in each country in 2013
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/WorldUniversity.csv')
df_2013 = df[df['year']== 2013]
average_score = df_2013.groupby('country')['score'].mean()
country = average_score.index
score = average_score.values
plt.figure(figsize = (10,8))
plt.bar(country , score , edgecolor = 'black' , color = 'skyblue')
plt.xlabel('Country')
plt.xticks(rotation= 45)
plt.ylabel('Average_score')
plt.title('Average scores for universities in each country in 2013')
plt.show()
```

```python
#17)	How can you use seaborn to create a swarm plot to visualize the distribution of scores among universities in 2014?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/WorldUniversity.csv')
df_2014 = df[df['year']== 2014]
plt.figure(figsize = (10,8))
sns.swarmplot(data = df_2014 , x = 'score', y = 'institution' )
plt.xlabel('score')
plt.ylabel('Institution')
plt.title('Distribution of scores among universities in 2014')
plt.tight_layout()
plt.show()
```

```python
#18)	Create a treemap using Altair to represent the proportion of universities in each country in 2012.

import pandas as pd
import altair as alt

df = pd.read_csv('/content/WorldUniversity.csv')
df_2012 = df[df['year']== 2012]
universities_count_country = df_2012['country'].value_counts().reset_index()
universities_count_country.columns = ['country', 'count']

tree_map = alt.Chart(universities_count_country).mark_rect().encode(
    x= 'count',
    y = 'country',
    color = 'count',
    tooltip = ['country' , 'count']
).properties(
    height = 400,
    width = 800,
    title = ('proportion of universities in each country in 2012')
)
tree_map
```

```python
#19)	How can you use matplotlib to create a histogram showing the distribution of scores for all universities in 2014?
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/WorldUniversity.csv')
df_2014 = df[df['year'] == 2014]
plt.hist(df_2014['score'], bins= 20 , edgecolor="black", color = "blue")
plt.xlabel('score')
plt.ylabel('Frequency')
plt.title('Distribution of scores for all universities in 2014')
plt.show()
```
